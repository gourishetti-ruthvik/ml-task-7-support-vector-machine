```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/breast-cancer.csv") #Loading dataset
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Drop irrelevant 'id' columnn
```

```python
df.drop(columns='id', inplace=True)
```

```python
# Mapping target column 'diagnosis' to binary
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

```python
df = df.dropna(subset=['diagnosis']) #Dropping rows where diagnosis is NaN
```

```python
# Selecting only required features
features = ['radius_mean', 'concave points_mean']
X = df[features]
y = df['diagnosis']
```

```python
#Normalizing the features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Training the SVM(Support Vector Machine) with Linear Kernel
```

```python
from sklearn.svm import SVC
```

```python
#Linear SVM
svm_linear = SVC(kernel='linear', C=1.0)
svm_linear.fit(X_scaled, y)
```

```python
#Visualizing the Decision Boundary using 2D data
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Plotting (radius_mean and texture_mean)
```

```python
X_plot = X_scaled[:, :2] #Taking first two features for 2D visualization
```

```python
def plot_decision_boundary(clf, X, y, title):
    h=0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() +1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), np.zeros_like(xx.ravel())])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.4, cmap='coolwarm')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
    plt.xlabel('radius_mean')
    plt.ylabel('texture_mean')
    plt.title(title)
    plt.show()

plot_decision_boundary(svm_linear, X_scaled, y, "SVM with Linear Kernel")
```

```python
#Training with RBF kernal and Hyperparameter Tuning
```

```python
svm_rbf = SVC(kernel='rbf', C=10, gamma=0.1)
svm_rbf.fit(X_scaled, y)
# Re-using same plot 
plot_decision_boundary(svm_rbf, X_scaled, y, "SVM with RBF Kernel (C=10, gamma=0.1)")
```

```python
#Cross-Validation Evaluation
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#Evaluating Linear kernel
linear_scores = cross_val_score(svm_linear, X_scaled, y, cv=5)
print(f"Linear Kernel Accuracy : {linear_scores.mean():.3f}")
```

```python
#Evaluating RBF Kernel
rbf_scores = cross_val_score(svm_rbf, X_scaled, y, cv=5)
print(f"RBF Kernel Accuracy : {rbf_scores.mean():.3f}")
```
